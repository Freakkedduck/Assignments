### 1. What is name of car that has the highest horsepower?  

```python
#your code here
df = pd.read_csv('Cars.csv')
df['horsepower'] = pd.to_numeric(df['horsepower'], errors='coerce')
max_hp = df['horsepower'].max()
car_with_max_hp = df[df['horsepower'] == max_hp]['name'].values
print("Car(s) with the highest horsepower:", car_with_max_hp)
print("Maximum horsepower:", max_hp)
```

### 2. How many cars have mpg ≥ 35?

```python
#your code here
count = (df['mpg'] >= 35).sum()
print("Number of cars with mpg ≥ 35:", count)
```

### 3. What is the most common origin for cars with horsepower > 100 and weight < 3000?

```python
#your code here
df['horsepower'] = pd.to_numeric(df['horsepower'], errors='coerce')
filtered = df[(df['horsepower'] > 100) & (df['weight'] < 3000)]
most_common_origin = filtered['origin'].mode().iloc[0]
print("Most common origin:", most_common_origin)
```

### 4. What is the mean acceleration of cars from Japan? (rounded to 2 decimals)

```python
#your code here
mean_acc_japan = df[df['origin'] == 'japan']['acceleration'].mean()
print("Mean acceleration (Japan):", round(mean_acc_japan, 2))
```

### 5. Which year had the highest average mpg?

```python
#your code here
avg_mpg_by_year = df.groupby('model_year')['mpg'].mean()
best_year = avg_mpg_by_year.idxmax()
print("Year with highest average mpg:", best_year)
```

### Find the car (or cars) with the best ratio of horsepower to weight among all cars that also have above-median mpg.

```python
#your code here
df['horsepower'] = pd.to_numeric(df['horsepower'], errors='coerce')
median_mpg = df['mpg'].median()
above_median = df[df['mpg'] > median_mpg].copy()
above_median['hp_weight_ratio'] = above_median['horsepower'] / above_median['weight']
max_ratio = above_median['hp_weight_ratio'].max()
best_cars = above_median[above_median['hp_weight_ratio'] == max_ratio]['name'].values
print("Car(s) with best hp/weight ratio (above-median mpg):", best_cars)
```

### Design a multi-line plot using Matplotlib or Seaborn that shows the evolution of average mpg over the years, separately for each origin

```python
#your code here
import seaborn as sns
avg_mpg_origin_year = df.groupby(['model_year', 'origin'])['mpg'].mean().reset_index()
plt.figure(figsize=(10,6))
sns.lineplot(data=avg_mpg_origin_year, x='model_year', y='mpg', hue='origin', marker='o')
plt.title('Average MPG Over Years by Origin')
plt.xlabel('Model Year')
plt.ylabel('Average MPG')
plt.legend(title='Origin')
plt.tight_layout()
plt.show()
```

### Create a Seaborn scatterplot (or PairGrid) where:

X = horsepower

Y = weight

Color by: origin

Size by: mpg

Hue order = ['japan', 'europe', 'usa']

Add meaningful plot titles and axis titles.

```python
#your code here
df = pd.read_csv('Cars.csv')
df['horsepower'] = pd.to_numeric(df['horsepower'], errors='coerce')
hue_order = ['japan', 'europe', 'usa']
plt.figure(figsize=(12, 7))
scatter = sns.scatterplot(
    data=df,
    x='horsepower',
    y='weight',
    hue='origin',
    size='mpg',
    sizes=(40, 400),
    palette='Set1',
    hue_order=hue_order,
    alpha=0.7,
    edgecolor='w',
    linewidth=0.5
)

plt.title('Horsepower vs Weight of Cars\n(Size = MPG, Color = Origin)', fontsize=16)
plt.xlabel('Horsepower', fontsize=14)
plt.ylabel('Weight (lbs)', fontsize=14)
plt.legend(title='Origin', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

### We define a “consistent” car model as one that was produced over multiple years and had very low variation in mpg across those years (standard deviation < 1.0).

Tasks:

Identify car names that appear in more than one model_year.

For each such name, compute the standard deviation of mpg across years.

Return the car(s) with the lowest variation in mpg, among those with at least 2 appearances and std(mpg) < 1.0.

Report the model name(s), number of appearances, and the average mpg.

Bonus:
Sort the result by number of appearances (descending), then mpg (descending).

```python
#your code here
df = pd.read_csv('Cars.csv')
name_year_counts = df.groupby('name')['model_year'].nunique()
multi_year_names = name_year_counts[name_year_counts > 1].index
grouped = df[df['name'].isin(multi_year_names)].groupby('name').agg(
    n_years=('model_year', 'nunique'),
    n_appearances=('name', 'count'),
    mpg_std=('mpg', 'std'),
    mpg_mean=('mpg', 'mean')
)
consistent = grouped[(grouped['mpg_std'] < 1.0) & (grouped['n_appearances'] >= 2)]
consistent = consistent.sort_values(by=['n_appearances', 'mpg_mean'], ascending=[False, False])
consistent = consistent.reset_index()
print(consistent[['name', 'n_appearances', 'mpg_mean']])
```
